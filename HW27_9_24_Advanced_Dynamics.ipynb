# Problem 9.24

```python
from sympy import *
from sympy.plotting import plot
from sympy import symbols
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
```

```python
t = var('t')
```

```python
x0, v0x, v0y, omega = symbols('x0 v0x v0y omega')
```

```python
omega = 1
x0 = 1
```

```python
x = Function('x')(t)
x = (x0 + v0x*t)*cos(omega*t) + (v0y+omega*x0)*t*sin(omega*t)
y = Function('y')(t)
y = -(x0 + v0x*t)*sin(omega*t)+ t*(v0y+omega*x0)*cos(omega*t)
```

## $v_0 = (0,1)$

```python
plot_parametric((x.subs({v0x:0, v0y:1}), y.subs({v0x:0, v0y:1})), (t, 0, 25), xlabel = 'x', ylabel = 'y');
```

# $v_0 = (0,0)$

```python
plot_parametric((x.subs({v0x:0, v0y:0}), y.subs({v0x:0, v0y:0})), (t, 0, 25), xlabel = 'x', ylabel = 'y');
```

# $v_0 = (0,-1)$

```python
plot_parametric((x.subs({v0x:0, v0y:-1}), y.subs({v0x:0, v0y:-1})), (t, 0, 25), xlabel = 'x', ylabel = 'y');
```

# $v_0 = (-0.5,-0.5)$

```python
plot_parametric((x.subs({v0x:-0.5, v0y:-0.5}), y.subs({v0x:-0.5, v0y:-0.5})), (t, 0, 25), xlabel = 'x', ylabel = 'y');
```

# $v_0 = (-0.7,-0.7)$

```python
plot_parametric((x.subs({v0x:-0.7, v0y:-0.7}), y.subs({v0x:-0.7, v0y:-0.7})), (t, 0, 25), xlabel = 'x', ylabel = 'y');
```

# $v_0 = (0,-0.1)$

```python
plot_parametric((x.subs({v0x:0, v0y:-0.1}), y.subs({v0x:0, v0y:-0.1})), (t, 0, 25), xlabel = 'x', ylabel = 'y');
```

Some intereseting features include the the closed orbit in the $v_0 = (0,-1)$ case. In all cases, we see sinusoidal behavior over time.
