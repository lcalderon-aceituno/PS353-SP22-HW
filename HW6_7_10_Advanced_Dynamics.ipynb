# 7.10 A particle is confined to move on the surface of a circular cone with its axis on the z axis, vertex at the origin, and a half-angle $\alpha$. Write down the equations that give you the three Cartesian coordinates of the particle in terms of the generalized coordinates $(\rho, \phi)$ and vice versa.

```python
from sympy import *
from sympy import symbols
from sympy import acot
```

20220217_234239.jpg

Based on the picture above, we get the following relationships:

```python
rho, phi, x, y, alpha = symbols('rho, phi, x, y, alpha') 
```

```python
Eq(S('x'),rho*cos(phi))
```

```python
Eq(S('y'),rho*sin(phi))
```

```python
Eq(S('z'),rho*cot(phi))
```

```python
Eq(S('rho'),x**2+y**2)
```

```python
Eq(S('alpha'),acot(z/(x**2+y**2)))
```
