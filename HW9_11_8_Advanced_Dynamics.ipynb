## part b
equilibrium positions are $x_1(0) = x_2(0) = 0$, but we now have $v_1(0) = -v_2(0)$

```python
B1 = 0
B2 = 0
v01 = 1
v02 = -v01
w2 = sqrt(3)
C1 = 0
C2 = -1/w2
```

```python
def x1(t):
  return (B1*cos(w1*t)+C1*sin(w1*t)+(B2*cos(w2*t)+C2*sin(w2*t)))
```

```python
def x2(t):
  return (B1*cos(w1*t)+C1*sin(w1*t)+(B2*cos(w2*t)-C2*sin(w2*t)))
```

```python
var('t') # create variable t
```

```python
fig = plot(x1(t), (t, 0, 30), show = False, label = "x1")
fig2 = plot(x2(t), (t, 0, 30), show = False, label = "x2", line_color = "purple")
fig.extend(fig2)
fig.legend = true
fig.show()
```
