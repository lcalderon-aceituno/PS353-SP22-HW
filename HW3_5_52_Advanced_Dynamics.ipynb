## initialize parameters

```python
tau0 = 1
omega0 = 2*pi/tau0
beta = 0.1
f0 = 1
```

## driving function

```python
var('t')
f = Piecewise( (0, t < -delta_tau/2), (0, t > +delta_tau/2), (fmax, True))
```

```python
tau = 1
fmax = 1
delta_tau = 0.25
omega = 2*pi/tau
f0 = integrate((1/tau)*f, (t, -tau/2, tau/2)) # calculate f0
```

```python
a0 = f0/omega0**2 # calculate a0

# define a(n)
def a(n):
  return fn(n)/sqrt((omega0**2 - n**2*omega**2)**2 + 4*beta**2*n**2*omega**2)

# define delta(n)
def delta(n):
  return atan2(2*beta*n*omega, omega0**2 - n**2*omega**2)

# define x(t)
def x(nmax, t):
  sym_x = a0 + Sum(a(n)*cos(n*omega*t-delta(n)), (n, 1, nmax))
  return sym_x.doit()

def fn(n):
  return integrate(2/tau*cos(n*omega*t)*f, (t, -tau/2, tau/2))
```

## x(t) calculations

### $\tau = 1$

```python
var('n', integer=True, positive=True)
fig1 = plot(x(6, t), (t, 0, 5*tau), line_color = 'red', label='n = 6', show=False, adaptive=False)
fig1.legend = True
fig1.show()
```
