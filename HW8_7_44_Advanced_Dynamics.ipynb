```python
import numpy as np
from scipy import integrate
import matplotlib.pyplot as plt
```

## part a

```python
g = 1
l = 1
w = 0.1
R = 0.2
```

```python
def phi(u, t):
  return (u[1], -g*np.sin(u[0])+w**2*R*np.cos(u[0] - w*t)) # equation of motion from problem 7.29
```

```python
phi0 = [0.2, 0] # initial conditions
time = np.linspace(0, 20, 1000)
soln, garbage = integrate.odeint(phi, phi0, time).T
```

```python
# plot solution
plt.plot(time, soln);
plt.xlabel('time (s)');
plt.ylabel('ɸ (rad)');
```

## part b

```python
longtime = np.linspace(0, 500, 100000)
longsoln, garbage = integrate.odeint(phi, phi0, longtime).T
```

```python
# plot solution
plt.plot(longtime, longsoln);
plt.xlabel('time (s)');
plt.ylabel('ɸ (rad)');
```

In the graph above, I looked at a longer period than suggested in the text (over 500 seconds of motion rather than 100), as it makes it easier to see how the amplitude of the oscillations grows and shrinks periodically. The period of these small oscillations is about $65$ seconds.
